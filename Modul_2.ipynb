```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub
import os

# 1. Load dataset
path = kagglehub.dataset_download("iabhishekofficial/mobile-price-classification")
df = pd.read_csv(path + '/train.csv')

# 2. Ambil fitur
X = df[['ram', 'battery_power']].values
y = df['price_range'].values

# 3. Softmax
def softmax(z):
    z = z - np.max(z)
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z)

# 4. Inisialisasi parameter
num_classes = 4
num_features = 2

W = np.random.randn(num_classes, num_features)
b = np.zeros(num_classes)

lr = 0.0001
epochs = 5

# 5. Training SGD (biar ada modelnya)
for epoch in range(epochs):
    for i in range(len(X)):
        xi = X[i]
        yi = y[i]

        z = np.dot(W, xi) + b
        probs = softmax(z)

        for c in range(num_classes):
            target = 1 if c == yi else 0
            error = probs[c] - target

            W[c] -= lr * error * xi
            b[c] -= lr * error

# 6. Buat grid
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 100),
    np.linspace(y_min, y_max, 100)
)

# 7. Prediksi grid
Z = np.zeros(xx.shape)

for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        point = np.array([xx[i, j], yy[i, j]])
        z = np.dot(W, point) + b
        probs = softmax(z)
        Z[i, j] = np.argmax(probs)

# 8. Visualisasi
plt.figure(figsize=(6,5))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')

plt.xlabel('RAM')
plt.ylabel('Battery Power')
plt.title('Decision Boundary')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = (X - X.mean(axis=0)) / X.std(axis=0)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

Z = np.zeros(xx.shape)

for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        point = np.array([xx[i, j], yy[i, j]])
        z = np.dot(W, point) + b
        probs = softmax(z)
        Z[i, j] = np.argmax(probs)

plt.figure(figsize=(7,6))

plt.contourf(xx, yy, Z, alpha=0.3)

plt.contour(xx, yy, Z, linewidths=1)

plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')

plt.xlabel('RAM (Normalized)')
plt.ylabel('Battery Power (Normalized)')
plt.title('Decision Boundary (Improved Visualization)')

plt.show()
```
